```python
import requests
from google.protobuf.internal.decoder import _DecodeVarint32

# API URL
url = "https://studentapi.api.sparxmaths.uk/sparx.swworker.v1.Sparxweb/GetPackageData"

# Required headers
headers = {
    "Authorization": "Bearer YOUR_VALID_AUTH_TOKEN",
    "Content-Type": "application/grpc-web+proto",
    "Origin": "https://www.sparxmaths.uk",
    "Referer": "https://www.sparxmaths.uk/",
    "User-Agent": "Mozilla/5.0 (Linux; Android 6.0; Nexus 5 Build/MRA58N) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/131.0.0.0 Mobile Safari/537.36",
    "x-grpc-web": "1",
}

# Required cookies
cookies = {
    "live-resolver-school": "huntingtonschool",
    "sparxweb_csrf": "e345a340-a9eb-4976-9c67-ed60ba5eaf15",
    "live_ssoprovider_session": "VALID_SESSION_COOKIE",
}

# Actual payload (binary)
payload = b"\x00\x00\x00\x00\x12\x24c\xf4\x1b\x64\xb7\xff\x4e\x40\x16\xa5\x28\x08\x71\xa7\x64\xc4\x04"

# Send the POST request
response = requests.post(url, headers=headers, cookies=cookies, data=payload)

# Decode gRPC response
if response.status_code == 200:
    data = response.content
    index = 0
    print(f'len(data):{len(data)}')
    while index < len(data):
        msg_length, new_index = _DecodeVarint32(data, index)
        index = new_index
        message = data[index: index + msg_length]
        print("Decoded Message:", message.decode("utf-8", errors="ignore"))
        index += msg_length
else:
    print(f"Error {response.status_code}: {response.text}")
```
